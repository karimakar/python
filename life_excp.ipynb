```python
# import our packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib.image as mpimg
import sklearn.linear_model
```

```python
# load the data
df = pd.read_csv('life_happy.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# preapre our data
X=np.array(df['GDP per capita']).reshape(-1,1)
y=np.array(df['Life Satisfaction']).reshape(-1,1)
```

```python
# vis step
df.plot(kind='scatter',x='GDP per capita',y='Life Satisfaction')
```

```python
# use seabron
sns.regplot(x='GDP per capita',y='Life Satisfaction',data=df)
```

```python
# select thr model
model = sklearn.linear_model.LinearRegression()
```

```python
# train the model
model.fit(X,y)
```

```python
# make prediction
x_new = [[22798]]
model.predict(x_new)
```
